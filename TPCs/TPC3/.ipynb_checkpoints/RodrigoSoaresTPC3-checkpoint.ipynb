### Introdução
Vamos começar por considerar a equação de advecção:
$$u_t(x, t) + c u_x(x, t) = 0,
$$ em que podemos vir a considerar a velocidade $c$ como dependendo do ponto (sistema inomogéneo).

Um método comum de obter a solução de PDEs dependentes do tempo é o chamado _método das linhas_ (MOL). Para aplicar este método, definimos uma grelha de pontos no domínio. Podemos por exemplo usar uma grelha de N pontos equiespaçados, e o método de Fourier, tomando como domínio o intervalo $[0, 2\pi]$ (deste modo, com condições fronteira apropriadas, podemos simular um sistema aberto). Nestes pontos da grelha obtemos um sistema de $N$ equações acopladas da forma:
$$ u_t(x_j, t) + c u_x(x_j, t) = 0, \quad \quad j=0, \ldots, N-1.
$$

Seja $U$ o vector cujas componentes são os valores $u(x_j, t)$. Denotemos a transformada de Fourier (no espaço) de $U$ como:
$$ \mathcal{F}(U)(t) \equiv \hat{u}(k, t),
$$
e podemos definir a sua inversa $\mathcal{F}^{-1}$ de um modo semelhante.

Usando a aproximação pseudo-espectral no espaço leva ao sistema de equações diferenciais ordinárias (ODE):
$$U_t + \vec{c} \mathcal{F}^{-1} (\vec{k} \mathcal{F}(U) ) = 0.
$$
Estas podem ser resolvidas por um método integrador comum (Euler, leap-frog, Runge-Kutta,...).

```python
def rk4_nD(func: callable, a: float, b: float, r: np.ndarray, h: float):
    num_steps = int((b - a) / h)
    num_vars = r.shape[0]

    t_values = np.linspace(a, b, num_steps + 1, dtype=np.float64)
    r_values = np.zeros((num_vars, num_steps + 1), dtype=np.complex128)
    
    r_values[:, 0] = r.astype(np.complex128)

    for i in range(num_steps):
        t = t_values[i]
        k1 = func(r, t)
        k2 = func(r + h / 2 * k1, t + h / 2)
        k3 = func(r + h / 2 * k2, t + h / 2)
        k4 = func(r + h * k3, t + h)
        
        r += h / 6 * (k1 + 2 * k2 + 2 * k3 + k4)
        
        r_values[:, i + 1] = r

    return t_values, r_values
```

```python
N = 100
Nt = 150
xx = np.linspace(0, 2 * np.pi, N)
c = 0.2 + np.sin(xx-1)**2
u0 = np.exp(-100 * (xx-1)**2)

def rhs1(u_hat, t):
    ik = 1j * np.fft.fftfreq(N, xx[1] - xx[0]) * 2 * np.pi
    u_x = np.fft.ifft(ik * u_hat)
    return -np.fft.fft(c * u_x)

tt, u_hat = rk4_nD(rhs1, 0, 8, np.fft.fft(u0), 8 / (Nt - 1))

u = np.fft.ifft(u_hat.T).real

X, T = np.meshgrid(xx, tt)

fig = plt.figure(figsize = (12, 8))
ax = fig.add_subplot(111, projection="3d")
ax.plot_wireframe(X,T, u)
plt.xlabel('x')
plt.ylabel('t')
plt.title('u(x, t)')
plt.tight_layout()
plt.show()
```

### b)

b) Desenvolva um código que permita obter a evolução no tempo desta condição inicial, usando um método pseudo-espectral (com polinómios de Chebyshev! Porque não pode/deve escolher Fourier?). Considere o caso $a=1$ e $b=0.2$. Escolha $x_0$ (que faça sentido). (Sugestão: o cap. 13 do livro de L. Trefethen _Spectral Methods with Matlab_ aborda o tratamento de condições fronteira inomogéneas))

$u(x, t) = v(x, t) + w(x)$

$ u(x, 0) = \begin{cases} a & \text{  se } x < x_0 \\
                             b & \text{  se } x > x_0.
\end{cases}
$

$u(0, t) = a$, $u(L, t) = b$.

Escolhemos $v$ tal que $v(0, t) = v(L, t) = 0$ de modo a ser mais fácil usar o método pseudo-espectral com Polinómios de Chebyshev.

$w(x) = a + x\left(\frac{b-a}{L}\right)$ tal que $w(0) = a$ e $w(L) = b$.

Substituindo na equação:
$$
v_t = - (v + w)\left(v_x + \frac{b-a}{L}\right)
$$

Não devemos escolher Fourier porque não temos condições fronteira periódicas: em $x=0$ temos $u_0 = a$ e em $x=L$ temos $u_0 = b$.

O domínio dos polinómios de Chebyshev é $[-1, 1]$, mas o nosso domínio é $[0, L]$.

$$
x = \frac{L}{2}(1 + \zeta)
$$

$$
\zeta_j = \cos\left(\frac{\pi j}{N}\right), j = 0, ..., N
$$

$$
\frac{d}{d\zeta} = \frac{d}{dx}\frac{dx}{d\zeta} = \frac{L}{2}\frac{d}{dx} \implies D_x = \frac{2}{L} D_\zeta
$$

```python
def cheb(N):
	if N==0: 
		D = 0.; x = 1.
	else:
		n = np.arange(0,N+1)
		x = np.cos(np.pi*n/N).reshape(N+1,1) 
		c = (np.hstack(( [2.], np.ones(N-1), [2.]))*(-1)**n).reshape(N+1,1)
		X = np.tile(x,(1,N+1))
		dX = X - X.T
		D = np.dot(c,1./c.T)/(dX+np.eye(N+1))
		D -= np.diag(np.sum(D.T,axis=0))
	return D, x.reshape(N+1)
```

```python
def rk4_nD(func: callable, a: float, b: float, r: np.ndarray, h: float, Nt: int):
    num_vars = r.shape[0]

    t_values = np.linspace(a, b, Nt, dtype=np.float64)
    r_values = np.zeros((num_vars, Nt), dtype=np.float64)
    
    r_values[:, 0] = r.astype(np.float64)

    for i in range(Nt-1):
        t = t_values[i]
        k1 = func(r, t)
        k2 = func(r + h / 2 * k1, t + h / 2)
        k3 = func(r + h / 2 * k2, t + h / 2)
        k4 = func(r + h * k3, t + h)
        
        r += h / 6 * (k1 + 2 * k2 + 2 * k3 + k4)
        
        r_values[:, i+1] = r

    return t_values, r_values
```

```python
a = 1
b = 0.2
x0 = 0.32898993 # Escolhi um ponto que pertence a xx
L = 1
N = 19
Nt = 100

dt = 1 / (Nt-1)

D_cheb, x_cheb = cheb(N-1)

Dx = (2/L) * D_cheb

xxint = L * (1 + x_cheb[1:-1]) / 2
xx = L * (1 + x_cheb) / 2

w = a + xx * (b - a) / L
wint = a + xxint * (b - a) / L

u0cheb = np.where(xx < x0, a, b)
v0 = u0cheb - w

def rhs2(vint, t):
    return -(vint + wint) * (Dx[1:-1, 1:-1] @ vint + (b - a) / L)

tt, vint = rk4_nD(rhs2, 0, 1, v0[1:-1], dt, Nt)

ucheb = np.zeros((N, Nt), dtype=np.float64)

W = np.zeros_like(ucheb)

for i in range(Nt):
    W[:, i] = w

ucheb[1:-1, :] = vint + W[1:-1, :]

ucheb[0, :] = b
ucheb[-1, :] = a
X, T = np.meshgrid(xx, tt)

fig = plt.figure(figsize = (12, 8))
ax = fig.add_subplot(111, projection="3d")
ax.plot_wireframe(X,T, ucheb.T)
plt.xlabel('x')
plt.ylabel('t')
plt.title('u(x, t)')
plt.tight_layout()
ax.set_zlim(0, 1)
plt.show()
```

```python
NN = np.arange(9, 20)

eps = 0.9

def Erro(N, Nt):
    dt = 1 / (Nt-1)
    
    D_cheb, x_cheb = cheb(N-1)
    
    Dx = (2/L) * D_cheb
    
    xxint = L * (1 + x_cheb[1:-1]) / 2
    xx = L * (1 + x_cheb) / 2
    
    w = a + xx * (b - a) / L
    wint = a + xxint * (b - a) / L
    
    u0cheb = np.where(xx < x0, a, b)
    v0 = u0cheb - w
    
    def rhs2(vint, t):
        return -(vint + wint) * (Dx[1:-1, 1:-1] @ vint + (b - a) / L)
    
    tt, vint = rk4_nD(rhs2, 0, 1, v0[1:-1], dt, Nt)
    
    ucheb = np.zeros((N, Nt), dtype=np.float64)
    
    W = np.zeros_like(ucheb)
    
    for i in range(Nt):
        W[:, i] = w
    
    ucheb[1:-1, :] = vint + W[1:-1, :]
    
    ucheb[0, :] = b
    ucheb[-1, :] = a

    erro = np.max(np.abs((ucheb[:, -1] - uanal(xx, 1))))
    #erro = np.sqrt(np.sum(ucheb[:, -1] - uanal(xx, 1))**2 * - np.diff(xx).mean())
    #erro = np.mean(np.abs(ucheb[:, -1] - uanal(xx, 1)))
    return erro, dt


dts = np.zeros_like(NN, dtype=np.float64)
Nts = np.zeros_like(NN, dtype=np.int32)

for i, N in enumerate(NN):
    Nt = 2 # ponto de partida
    while True:
        erro, dt = Erro(N, Nt)
        if erro <= eps:
            dts[i] = dt
            Nts[i] = Nt
            break
        Nt += 1

plt.figure(figsize = (12, 8))
plt.loglog(NN, dts, 'ko')
plt.xlabel('N')
plt.ylabel('dt')
plt.grid()
plt.show()
```

# Outro método

Ao olhar para os meus resultados não fiquei muito satisfeito, porque sabia que o erro estava grande também pelo facto de não poder usar N>19.

O que decidi fazer foi usar substituir o rk4 pelo solver do scipy e testar os diferentes métodos. O que me deu bem foi o BDF1 que é um Backward Euler.

https://en.wikipedia.org/wiki/Backward_Euler_method

#### Backward Euler

$$
\frac{\partial v}{\partial t} = F(v)
$$

---

$$
\frac{v^{n+1} - v^{n}}{\Delta t} = F (v^{n+1}) \Leftrightarrow v^{n+1} - v^{n} - \Delta t F(v^{n+1}) = 0
$$

Definimos $G(v^{n+1}) = v^{n+1} - v^{n} - \Delta t \cdot F(v^{n+1})$

Agora encontramos o zero de $G$ com o método de Newton–Raphson. https://en.wikipedia.org/wiki/Newton%27s_method

#### Método de Newton–Raphson
Definimos $y = v^{n+1}$

$$
y^{k+1} = y^{k} - \frac{G(y^{k})}{J_G(y^{k})}, \, \, \, G(y) = y - v^{n} - \Delta t \cdot F(y)
$$

------

$$
J_G = \frac{\partial G}{\partial y} = I - \Delta t \frac{\partial F}{\partial y}
$$

---

$$
F(y) = - (y + w)\left(D_x \,y + \frac{b-a}{L}\right)
$$

$$
\frac{\partial F}{\partial y} = - diag\left(D_x \,y + \frac{b-a}{L}\right) - diag(y + w) D_x
$$

---

$$
J_G(y^k) (y^{k+1} - y^{k}) = - G(y^{k})
$$

Penso que a razão deste método dar bem é o facto de ser implícito.

```python
N = 100
dt = 1 / N

Nt = int(1 + 1 / dt)
D_cheb, x_cheb = cheb(N-1)

Dx = (2/L) * D_cheb

xxint = L * (1 + x_cheb[1:-1]) / 2
xx = L * (1 + x_cheb) / 2

w = a + xx * (b - a) / L
wint = a + xxint * (b - a) / L

u0cheb = np.where(xx < x0, a, b)
v0 = u0cheb - w

def F(y):
    return -(y + wint) * (Dx[1:-1, 1:-1] @ y + (b - a) / L)

def JG(y):
    return np.eye(len(y)) + dt * (np.diag(Dx[1:-1, 1:-1] @ y + (b - a) / L) + np.diag(y + wint) @ Dx[1:-1, 1:-1])

def BDF(y, tol=1e-10, Nmax=20):
    ynew = y.copy()
    for _ in range(Nmax):
        G = ynew - y - dt * F(ynew)

        delta = np.linalg.solve(JG(ynew), -G)
        
        ynew += delta

        if np.max(np.abs(delta)) < tol:
            break
    return ynew

vint = np.zeros((len(v0[1:-1]), Nt))
vint[:, 0] = v0[1:-1]

for i in range(Nt-1):
    vint[:, i+1] = BDF(vint[:, i])

u = np.zeros((N, Nt), dtype=np.float64)

W = np.zeros_like(u)

for i in range(Nt):
    W[:, i] = w

u[1:-1, :] = vint + W[1:-1, :]

u[0, :] = b
u[-1, :] = a

X, T = np.meshgrid(xx, np.linspace(0, 1, Nt))

fig = plt.figure(figsize = (12, 8))
ax = fig.add_subplot(111, projection="3d")
ax.plot_wireframe(X,T, u.T)
plt.xlabel('x')
plt.ylabel('t')
plt.title('u(x, t)')
plt.tight_layout()
plt.show()
```
