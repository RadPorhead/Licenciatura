contém alguns segundos de um discurso famoso. Mas infelizmente o ficheiro foi corrompido com ruído, de tal modo que mal é perceptível o que é dito. O vosso objectivo é tentar melhorar o áudio o mais possível.

Numa primeira fase vamos tentar filtrar o sinal no domínio das frequências. Sabemos que o ficheiro tem uma taxa de amostragem de $44100\text{ Hz}$, e tem uma duração de cerca de $11\text{ s}$.

```python
Audio('audio1wn.wav')
```

#### Não consigo perceber o que é dito, mas é possível perceber que é um homem adulto a falar.

```python
fs, audio = wavfile.read('audio1wn.wav')
audio = audio.astype(np.float64)

N = len(audio)
dt = 1 / fs

freqs = np.fft.fftfreq(N, dt)
fft = np.fft.fft(audio)

plt.figure(figsize = (10, 6))

plt.subplot(211)
plt.plot(np.arange(0, N/fs, dt), audio)
plt.title('Áudio')
plt.xlabel('Tempo (s)')
plt.ylabel('Amplitude')

plt.subplot(212)
plt.plot(freqs[freqs>=0], np.abs(fft[freqs>=0]))
plt.title('Espectro')
plt.xlabel('k')
plt.ylabel(r'$|c_k|$')

plt.tight_layout()
plt.show()
```

#### Por análise do espectro é possível observar que ele é aproximadamente constante, o que pode indicar ruído branco.

### Alínea b)

b) Procure criar filtros no domínio da frequência para "limpar" o áudio. (não pode usar filtros de pacotes, tem que fazer os seus!)

#### https://www.researchgate.net/publication/389207542_Pitch_Imperfect_Detecting_Audio_Deepfakes_Through_Acoustic_Prosodic_Analysis

#### Neste artigo é dito que a frequência fundamental de um homem adulto é 125 Hz. 

#### Com isso em mente, tentei zerar todas as frequências à volta da frequência fundamental e dos seus harmónicos com uma dada margem.

#### Obtive o seguinte resultado:

```python
filtro1 = np.zeros(N, dtype=np.float64)

margem = 5
f1 = 125

while f1 <= fs/2:
    filtro1[(np.abs(freqs - f1) < margem) | (np.abs(freqs + f1) < margem)] = 1
    f1 += 125

output1 = np.fft.ifft(fft * filtro1).real

plt.plot(freqs[freqs>=0], filtro1[freqs>=0])
plt.xlabel('k')
plt.ylabel('Amplitude')
plt.title('Filtro 1')
plt.show()

Audio(output1, rate=fs)
```

#### Isto não funcionou porque a voz humana não contém só esta frequência e os seus harmónicos e esta frequência também está bem presente no ruído.

#### A minha outra ideia foi fazer uma média ponderada de um filtro passa-baixo com um passa-alto e ir variando as frequências de corte.

```python
fpa = 100
fpb = 1000

passa_baixo = 1 / (1 + (freqs / fpb)**2)
passa_alto = (freqs/fpa)**2 / (1 + (freqs/fpa)**2)

filtro2 = 0.7 * passa_baixo + 0.3 * passa_alto

output2 = np.fft.ifft(fft * filtro2).real

plt.plot(freqs[freqs>=0], filtro2[freqs>=0])
plt.xlabel('k')
plt.ylabel('Amplitude')
plt.title('Filtro 2')
plt.show()

Audio(output2, rate=fs)
```

#### Ainda não se consegue ouvir, mas notam-se melhorias.

### Alínea c)

c) Quando as características do ruído variam ao longo do tempo a TF por si só não ajuda muito. Nesse caso é mais útil usar a TF em janelas em que dividimos o array áudio original. Depois podemos criar um gráfico da grandeza dos coeficientes de Fourier em função da frequência e do tempo.
O algoritmo da transformada de Fourier de duração curta (TFDC, ou STDT em inglês) é o seguinte:

1. Tomar um pequeno segmento dos dados do sinal total:
  Selecionamos um segmento : precisamos encontrar o nosso segmento atual para processar a partir do conjunto geral de dados. Usamos o conceito de _janela deslizante_ para nos ajudar a visualizar o que está a acontecer. Os dados dentro da janela são o segmento atual a ser processado. 
O comprimento da janela permanece o mesmo durante o processamento dos dados, mas o deslocamento muda a cada etapa do algoritmo. Normalmente, ao processar o TFDC, a alteração no deslocamento será menor que o comprimento de uma janela, o que significa que a última janela e a janela atual se sobrepõem. Se definirmos o tamanho da janela e a percentagem de sobreposição, saberemos todas as informações de que precisamos sobre como a janela se move ao longo do processamento.
2. Multiplicar esse segmento por um _meio-cosseno_;
  Isto ajuda a aliviar um problema criado pela segmentação dos dados. Quando os dados são cortados em pedaços, as bordas fazem uma transição brusca que não existia antes. Multiplicar por uma função de _meio cosseno_ ajuda a "amortecer" o sinal no início e no fim da janela, de modo que as transições nas bordas não afetem a transformada de Fourier dos dados.
Há um benefício adicional em usar uma janela de _meio cosseno_. Pode ser demonstrado que com um fator de sobreposição de 50%, a soma das janelas é um valor constante de 1. Isso significa que multiplicar os nossos dados por esta função específica não introduz diferenças na amplitude do sinal original. Isto é particularmente importante se a forma de onda tiver algum tipo de unidade e quisermos que as frequências na transformada de Fourier representem corretamente o conteúdo de energia de cada frequência nessas unidades.
   O _meio-cosseno_ na forma discreta pode ser escrito como $\sin^2(\pi n/N)$ com $n=0, \ldots,N$.
4. Adicionar zeros até aos limites do segmento:
  Preenchemos o final do segmento atual com um número de zeros igual ao comprimento da janela (é sempre melhor colocar estes zeros simetricamente antes e depois dos dados). Ou, noutras palavras, o novo segmento será duas vezes maior que o segmento original. O motivo para isso é explicado mais tarde. 
5. Tomar a transformada de Fourier desse segmento e normalizar as energias às frequências positivas e negativas (usar teorema de Parseval, p.ex.), dividindo pelo número de amostras.
6. Combinar a energia das frequências negativas e positivas, e exibir o espectro de um dos lados (em geral o positivo!);
7. (pode representar o espectro resultante em dB, para se ver melhor os contrastes ->Não é essencial!);
8. apodizar (podar!) o sinal para remover o ruído até ao limite que não nos interessa (valores muito baixos são erro numérico), para aumentar o contraste de cores na parte que interessa.

Note que o salto lateral entre janelas consecutivas é em geral menor que o tamanho de cada janela. Como referido no ponto 2. há vantagens em tomar o salto como metade da janela. 

Um exemplo ilustrativo do significado dos termos está na figura abaixo.
Podem ver mais detalhes, se o desejarem, em https://en.wikipedia.org/wiki/Short-time_Fourier_transform.

<figure>
<img src="Short-time-Fourier-transform.jpg" style="width:80%">
<figcaption align = "center"> from DOI:10.3390/app10207208 </figcaption>
</figure>

Implemente este algoritmo.

```python
L = 512 # tamanho da janela
step = L//2 # hop length=overlap length, sobreposição de 50%

window = np.sin(np.pi * np.arange(L) / L)**2 # função window

Nw = int(np.floor((N - L) / step) + 1) # número de janelas
audio_pad = np.concatenate((audio, np.zeros(abs((Nw - 1) * step + L - N)))) # fazer STFT do áudio todo

stft = []

for i in range(Nw):
    start = i * step
    frame = audio_pad[start: start + L] * window
    frame_pad = np.pad(frame, (L//2, L//2), 'constant', constant_values = (0,0))
    spectrum = np.fft.fft(frame_pad) / (2 * L)
    stft.append(spectrum)

stft = np.array(stft)
mag, phase = np.abs(stft), np.angle(stft)

spectrogram = 20 * np.log10(mag[:, :L//2 + 1])

plt.figure(figsize = (10, 4))
plt.imshow(spectrogram.T, extent = [0, N/fs, 0, fs/2], aspect = 'auto', origin = 'lower', cmap = 'hot')
plt.title('Espectrograma')
plt.xlabel('Tempo (s)')
plt.ylabel('Frequência (Hz)')
plt.colorbar(label = 'dB')
plt.show()
```

### Alínea d)

d) Aplique esta função ao áudio original. Estude a distribuição de energia pelas frequências. Veja se pode implementar um filtro simples que lhe permita melhorar o sinal. (pode usar a densidade de potência espectral).

```python
PSD = np.mean(mag**2, axis = 0) # média da densidade de potência espectral em dB ao longo do tempo para cada freq

PSD /= np.max(PSD)

plt.figure(figsize = (10, 6))
plt.plot(np.fft.fftfreq(2*L, dt)[:L//2 + 1], PSD[:L//2 + 1], 'o-')
plt.xlabel('Frequência (Hz)')
plt.ylabel('PSD')
plt.title('Densidade de Potência Espectral (Normalizada)')
plt.show()
```

#### PSD varia muito pouco ao longo das frequências e é muito oscilatório.

```python
mask = mag > np.percentile(mag, 5, axis=0)

plt.figure(figsize = (10, 4))
plt.imshow(mask.T, extent = [0, N/fs, 0, fs/2], aspect = 'auto', origin = 'lower', cmap = 'gray')
plt.title('Máscara')
plt.xlabel('Tempo (s)')
plt.ylabel('Frequência (Hz)')
plt.colorbar(label = 'dB')
plt.show()
```

```python
output3 = np.zeros_like(audio_pad)
ww = np.zeros_like(audio_pad)

for i in range(Nw):
    start = i * step
    output3[start: start + L] += (np.fft.ifft(stft[i] * mask[i]).real)[L//2:L//2 + L]
    ww[start: start + L] += window

ww[ww == 0] = 1 #evita divisão por zero

output3 = (output3/ww)[:N]

Audio(output3, rate=fs)
```

#### Não foi possível filtrar melhor o áudio devido à natureza oscilatória do ruído.

### Alínea e)

e) Posteriormente foi possível encontrar um outro audio em que só existe ruído (_noise.wav_). Acredita-se que tenha sido gravado antes do discurso (como teste). Veja se consegue melhorar o áudio original aplicando a função de TFDC ao sinal, ao ruído , subtrai a transformada do ruído da do sinal (apenas as grandezas), inverte e recombina com a fase do sinal (que calculou e guardou antes, pois foi?) para reobter um novo áudio filtrado. Como se compara este resultado com o anterior?

```python
_, noise = wavfile.read('noise.wav')
noise = noise.astype(np.float64)
noise_pad = np.concatenate((noise, np.zeros(abs((Nw - 1) * step + L - N))))

stft_noise = []

for i in range(Nw):
    start = i * step
    frame = noise_pad[start: start + L] * window
    frame_pad = np.pad(frame, (L//2, L//2), 'constant', constant_values = (0,0))
    spectrum = np.fft.fft(frame_pad) / (2 * L)
    stft_noise.append(spectrum)

stft_noise = np.array(stft_noise)
mag_noise = np.abs(stft_noise)

C = 0.8 # parâmetro de ajuste
mag_final = np.maximum(mag - C * mag_noise, 0) # evita mag<0
final = mag_final * np.exp(1j * phase)
```

```python
output4 = np.zeros_like(audio_pad)
ww = np.zeros_like(audio_pad)

for i in range(Nw):
    start = i * step
    output4[start: start + L] += (np.fft.ifft(final[i]).real)[L//2:L//2 + L]
    ww[start: start + L] += window

ww[ww == 0] = 1 #evita divisão por zero

output4 = (output4/ww)[:N]

Audio(output4, rate=fs)
```

#### Os resultados são muito melhores do que os da alínea anterior.

#### Para C = 0.8 ouve-se sem ruído mas com menor amplitude.
#### Para C = 1 ouve-se com algum ruído mas com maior amplitude.
