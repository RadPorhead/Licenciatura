```python
def fatorial(n: int):
    if n == 0:
        return 1
    return n * fatorial(n - 1)

# Se usar floats:

def fatorial2(n: int):
    if n == 0:
        return 1.0
    return n * fatorial2(n - 1)

print(fatorial(200))
print()
print(fatorial2(200))
```

$$
f'(x) = 2x - 1 \Rightarrow f'(1) = 1
$$

```python
def f(x):
    return x * (x - 1)

def derivative(func: callable, x0: float, eps: float) -> float:
    return (func(x0 + eps) - func(x0)) / eps

epsilons = np.array([1e-2, 1e-4, 1e-6, 1e-8, 1e-10, 1e-12, 1e-14])

for i, eps in enumerate(epsilons):
    print(abs(derivative(f, 1, eps) - 1))
```

A partir de $\delta = 10^{-10}$, os resultados começam a ser piores, pois envolvem a subtração de dois números muito próximos, o que resulta em uma perda significativa de precisão devido ao arredondamento dos números em ponto flutuante.

```python
a = -1
b = 1
N_values = np.array([100, 2**24])

for i, N in enumerate(N_values):
    h = (b - a) / N
    x = np.linspace(a, b, N + 1)
    y = np.sqrt(1 - x**2)
    I = h * sum(y)
    print(abs(I - np.pi/2))
```
