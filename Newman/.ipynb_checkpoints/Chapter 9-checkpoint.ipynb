```python
def poisson2d(phi: np.ndarray, solve: np.ndarray, f: np.ndarray, h: float, target: float=1e-6, w: float=0.9, N_max: int=10000) -> np.ndarray:
    """
    Resolve a equa√ß√£o de Poisson a 2D `nabla^2 phi = f` numa grelha de pontos equidistantes usando o m√©todo de Gauss-Seidel com sobrerelaxa√ß√£o come√ßando com estimativa inicial `phi` e terminando quando o passo em cada ponto for menor que `target`.

    A matriz `solve` cont√©m True caso o ponto esteja na √°rea governada pela equa√ß√£o e cont√©m False caso esteja na fronteira.
    """

    # Dimens√µes do problema
    Nx, Ny = phi.shape

    erro = 1.0
    c = 0
    while erro > target and c < N_max:
        oldphi = np.copy(phi)
        
        for i in range(Nx):
            for j in range(Ny):
                if solve[i, j]:
                    phi[i, j] = (1 + w) * 0.25 * (phi[i+1, j] + phi[i-1, j] + phi[i, j+1] + phi[i, j-1] - f[i, j] * (h**2)) - w * phi[i, j]

        erro = np.max(abs(phi - oldphi))
        c += 1
    
    print(f'N√∫mero de itera√ß√µes: {c}')
    return phi


# Par√¢metros
L = 1.0
N = 100
h = L / N # Nota: Estou a considerar dx = dy

phi = np.zeros((N+1, N+1))
solve = np.full((N + 1, N + 1), True)

# Fonte
f = np.zeros((N+1, N+1))
for i in range(N + 1):
    for j in range(N + 1):
        if  20 <= i <= 40 and 20 <= j <= 40:
            f[i, j] = -1
        elif  60 <= i <= 80 and 60 <= j <= 80:
            f[i, j] = 1
        elif (i in [0, N]) or (j in [0, N]):
            solve[i, j] = False

sol = poisson2d(phi, solve, f, h)

plt.imshow(sol, origin = 'lower', cmap='viridis')
plt.colorbar(label='Potencial œÜ(x, y)')
plt.title('Distribui√ß√£o do Potencial El√©trico')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def laplace2d(phi: np.ndarray, solve: np.ndarray, target: float=1e-6, w: float=0.9, N_max: int=1000) -> np.ndarray:
    """
    Resolve a equa√ß√£o de Laplace a 2D `nabla^2 phi = 0` numa grelha de pontos equidistantes usando o m√©todo de Gauss-Seidel com sobrerelaxa√ß√£o come√ßando com estimativa inicial `phi` e terminando quando o passo em cada ponto for menor que `eps`.

    A matriz `solve` cont√©m True caso o ponto esteja na √°rea governada pela equa√ß√£o e cont√©m False caso esteja na fronteira.
    """

    # Dimens√µes do problema
    Nx, Ny = phi.shape

    erro = 1.0
    c = 0
    while erro > target and c <= N_max:
        oldphi = np.copy(phi)
        
        for i in range(Nx):
            for j in range(Ny):
                if solve[i, j]:
                    phi[i, j] = (1 + w) * 0.25 * (phi[i+1, j] + phi[i-1, j] + phi[i, j+1] + phi[i, j-1]) - w * phi[i, j]

        erro = np.max(abs(phi - oldphi))
        c += 1
    
    print(f'N√∫mero de itera√ß√µes: {c}')
    return phi

# Par√¢metros
N = 100

phi = np.zeros((N+1, N+1))
solve = np.full((N + 1, N + 1), True)

# Condi√ß√µes do problema
for i in range(N + 1):
    for j in range(N + 1):
        if (i in [0, N]) or j == 0:
            solve[i, j] = False
        elif j == N:
            solve[i, j] = False
            phi[i, j] = 1

sol = laplace2d(phi, solve)

plt.imshow(sol.T, origin = 'lower', cmap='viridis')
plt.colorbar(label='Potencial œÜ(x, y)')
plt.title('Distribui√ß√£o do Potencial El√©trico')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Thermal diffusion in the Earth‚Äôs crust

A classic example of a diffusion problem with a time-varying boundary condition is the diffusion of heat into the crust of the Earth, as surface temperature varies with the seasons. Suppose the mean daily temperature at a particular point on the surface varies as:

$$
T_0(t) = A+B \,\sin(2\pi t / \tau)
$$

$
\tau = 365 \,\text{days}, \,A = 10 \celsius, B = 20 \celsius 
$

At a depth of $20\,m$ below the surface almost all annual temperature variation is ironed out and the temperature is, to a good approximation, a constant $11\, \celsius$ (which is higher than the mean surface temperature of $10\, \celsius$‚Äîtemperature increases with depth, due to heating from the hot core of the planet). The thermal diffusivity of the Earth‚Äôs crust varies somewhat from place to place, but for our purposes we will treat it as constant with value$\,D = 0.1\,m^2\,day^{‚àí1}$.

Write a program, or modify one of the ones given in this chapter, to calculate the temperature profile of the crust as a function of depth up to $20\,m$ and time up to $10\,$ years. Start with temperature everywhere equal to $10\, \celsius$, except at the surface and the deepest point, choose values for the number of grid points and the time-step $h$, then run your program for the first nine simulated years, to allow it to settle down into whatever pattern it reaches. 

Then for the tenth and final year plot four temperature profiles taken at 3-month intervals on a single graph to illustrate how the temperature changes as a function of depth and time.

$$
T(depth = 20 m) = 11 \celsius
$$
$$
T_{m√©dia \,√† \,superf√≠cie} = 10 \celsius
$$
$$
D = 0.1\,m^2\,day^{‚àí1} 
$$

No in√≠cio: $T = 10 \celsius$ em todo lado exceto na superf√≠cie e no ponto mais profundo.

Queremos a temperatura da crosta em fun√ß√£o da profundidade at√© √† profundidade de $20 \,m$ e o tempo at√© $9 \,anos$.

```python
def T0(t):
    A = 10
    B = 20
    tau = 365
    return A + B * np.sin(2 * np.pi * t / tau)

def Difusion1D(Nx, dt, dx):
    D = 0.1
    F = D * dt / dx**2

    # Montar matriz para pontos internos: de 1 at√© Nx-2 (Nx-2 pontos)
    N = Nx - 2
    main_diag = np.ones(N) * (1 + F)
    off_diag = np.ones(N - 1) * (-0.5 * F)
    
    A = diags([main_diag, off_diag, off_diag], [0, -1, 1], format='csc')
    
    main_diag_B = np.ones(N) * (1 - F)
    off_diag_B = np.ones(N - 1) * (0.5 * F)
    B = diags([main_diag_B, off_diag_B, off_diag_B], [0, -1, 1], format='csc')
    
    return spla.factorized(A), B

Nx = 100

xx = np.linspace(0, 20, Nx)
dx = xx[1] - xx[0]
dt = 0.2    
D = 0.1
F = D * dt / dx**2

Nt = int(9 * 365 / dt) + 1

T = np.ones_like(xx) * 10.0

solveA, B = Difusion1D(Nx, dt, dx)

for frame in range(Nt):
    t = frame * dt

    # Impor condi√ß√µes de fronteira
    T[0] = T0(t)
    T[-1] = 11.0

    # Construir vetor do lado direito com corre√ß√£o nas bordas
    b = B @ T[1:-1]
    b[0] += 0.5 * F * (T[0] + T[0])  # contribui√ß√£o da borda esquerda
    b[-1] += 0.5 * F * (T[-1] + T[-1])  # contribui√ß√£o da borda direita

    T[1:-1] = solveA(b)

plt.figure(figsize = (12, 4))
plt.xlabel('Depth (m)')
plt.plot(xx, T, 'r--')
plt.ylabel('T (¬∫C)')
plt.title('Difus√£o T√©rmica na Crosta Terrestre')
plt.grid()
plt.show()
```

```python
Nt = int(10 * 365 / dt) + 1

space = int(3 * 30 /dt)

T = np.ones_like(xx) * 10.0

solveA, B = Difusion1D(Nx, dt, dx)

plt.figure(figsize = (12, 4))

for frame in range(Nt):
    t = frame * dt

    # Impor condi√ß√µes de fronteira
    T[0] = T0(t)
    T[-1] = 11.0

    # Construir vetor do lado direito com corre√ß√£o nas bordas
    b = B @ T[1:-1]
    b[0] += 0.5 * F * (T[0] + T[0])  # contribui√ß√£o da borda esquerda
    b[-1] += 0.5 * F * (T[-1] + T[-1])  # contribui√ß√£o da borda direita


    T[1:-1] = solveA(b)
    if t == 365 * 9 or t == 365 * 9 + 3*30 or t == 365 * 9 + 6*30 or t == 365 * 9 + 9*30:
        plt.plot(xx, T, label = f'{(t-365*9)/30:.4f} meses')

plt.xlabel('Depth (m)')
plt.ylabel('T (¬∫C)')
plt.title('Difus√£o T√©rmica na Crosta Terrestre')
plt.grid()
plt.legend()
plt.show()
```

# Exemplo Newman

```python
start = time.time()

T0 = 20.0
Tc = 0.0
Th = 50.0
D = 4.25e-2

tt = [0.01, 0.1, 0.4, 1, 10]

def Difusion1D(Nx, dt, dx):
    D = 4.25e-2
    F = 0.49

    # Montar matriz para pontos internos: de 1 at√© Nx-2 (Nx-2 pontos)
    N = Nx - 2
    main_diag = np.ones(N) * (1 + F)
    off_diag = np.ones(N - 1) * (-0.5 * F)
    
    A = diags([main_diag, off_diag, off_diag], [0, -1, 1], format='csc')
    
    main_diag_B = np.ones(N) * (1 - F)
    off_diag_B = np.ones(N - 1) * (0.5 * F)
    B = diags([main_diag_B, off_diag_B, off_diag_B], [0, -1, 1], format='csc')
    
    return spla.factorized(A), B

Nx = 100
xx = np.linspace(0, 1, Nx)
dx = xx[1] - xx[0]
F = 0.49
dt = F * dx**2 / D
Nt = int(10 / dt) + 2
 
T = np.full(Nx, T0, float)
T[0] = Th
T[-1] = Tc
    
solveA, B = Difusion1D(Nx, dt, dx)

eps = 1e-3
i = 0
for frame in range(Nt):
    t = frame * dt

    # Impor condi√ß√µes de fronteira
    T[0] = Th
    T[-1] = Tc

    # Construir vetor do lado direito com corre√ß√£o nas bordas
    b = B @ T[1:-1]
    b[0] += 0.5 * F * (T[0] + T[0])  # contribui√ß√£o da borda esquerda
    b[-1] += 0.5 * F * (T[-1] + T[-1])  # contribui√ß√£o da borda direita

    T[1:-1] = solveA(b)

    if abs(t - tt[i]) < eps:
        plt.plot(xx, T, label = f't = {t} s')
        i += 1

plt.xlabel('Depth (cm)')
plt.ylabel('T (¬∫C)')
plt.grid()
plt.legend()
plt.show()

end = time.time()
print(end-start)
```

# Onda

Consider a piano string of length L, initially at rest. At time t = 0 the string is struck by the piano hammer a distance d from the end of the string.

The string vibrates as a result of being struck, except at the ends, $x = 0$ and $x = L$, where it is held fixed.

Write a program that solves the wave equation for the case $v = 100\,m\,s^{‚àí1}$, with the initial condition that $\phi(x) = 0$ everywhere but the velocity $œà(x)$ is nonzero, with profile:

$$
\psi(x) = C \frac{x(L-x)}{L^2} \exp\left(-\frac{(x-d)^2}{2\sigma^2}\right)
$$
where $L = 1\,m, d = 10\,cm, C = 1\,m\,s^{‚àí1}, \sigma = 0.3\,m$.

```python
def psi(x, d = 0.1, C = 1, sigma = 0.3):
    return C * x * (1 - x) * np.exp(- (x - d)**2 / (2 * sigma**2))

Nx = 50
T = 0.02 #s

N = Nx - 2

xx = np.linspace(0, 1, Nx)
dx = xx[1] - xx[0]

c = 1.0
v = 100
dt = c * dx / v

Nt = int(T / dt)

mainA = np.ones(N) * (4 / c**2 + 2)
offA = np.ones(N-1) * -1

mainB = np.ones(N) * (8 / c**2 - 4)
offB = np.ones(N-1) * 2

mainC = np.ones(N) * -(4/c**2 + 2)
offC = np.ones(N-1)

A = diags([offA, mainA, offA], offsets=[-1, 0, 1], format='csc')
B = diags([offB, mainB, offB], offsets=[-1, 0, 1], format='csc')
C = diags([offC, mainC, offC], offsets=[-1, 0, 1], format='csc')

solveA = spla.factorized(A)

new = np.zeros(Nx)
old = np.zeros(Nx)
now = np.zeros(Nx)
now[1:-1] = solveA((dt * 4/c**2) * psi(xx[1:-1]))
now[0] = 0
now[-1] = 0

fig, ax = plt.subplots()
line, = ax.plot(xx, now)
ax.set_ylim(-0.001, 0.001)
ax.set_xlim(0, 1)
ax.grid()

# Fun√ß√£o de anima√ß√£o que atualiza e plota "new"
def update(frame):
    global old, now, new
    new[1:-1] = solveA(B @ now[1:-1] + C @ old[1:-1])
    new[0] = 0
    new[-1] = 0
    old[:], now[:] = now[:], new[:]
    line.set_ydata(new)
    ax.set_title(f'{frame*dt:.4f}s')
    return line,

ani = FuncAnimation(fig, update, frames=Nt, interval=1, blit=True)
ani.save("onda.gif", writer=PillowWriter(fps=30))
plt.show()
```

# Extremos Fechados

u(x, 0) = \sin(k_n x)

u'(x, 0) = 0

```python
Nx = 100

N = Nx - 2
L = 1
T = 2
xx = np.linspace(0, L, Nx)

c = 1.0
v = 1
dx = xx[1] - xx[0]
dt = c * dx / v

Nt = int(T / dt) + 1

mainA = np.ones(N) * (4 / c**2 + 2)
offA = np.ones(N-1) * -1

mainB = np.ones(N) * (8 / c**2 - 4)
offB = np.ones(N-1) * 2

mainC = np.ones(N) * -(4/c**2 + 2)
offC = np.ones(N-1)

A = diags([offA, mainA, offA], offsets=[-1, 0, 1], format='csc')
B = diags([offB, mainB, offB], offsets=[-1, 0, 1], format='csc')
C = diags([offC, mainC, offC], offsets=[-1, 0, 1], format='csc')

solveA = spla.factorized(A)

n = 1
k = n * np.pi / L

new = np.zeros(Nx)
old = np.sin(k * xx)
now = np.zeros(Nx)
now[1:-1] = spla.spsolve(A-C, B@old[1:-1])
now[0] = 0
now[-1] = 0

# Fun√ß√£o de atualiza√ß√£o para a anima√ß√£o
for frame in range(Nt):
    new[1:-1] = solveA(B @ now[1:-1] + C @ old[1:-1])
    new[-1] = 0
    new[0] = 0
    old[:] = now[:]
    now[:] = new[:]

def analitica1(x, t):
    omega = v * k 
    return np.sin(k * x) * np.cos(omega * t)

tempo = frame * dt

plt.plot(xx, new)
plt.grid()
plt.show()

plt.plot(xx, np.abs(new - analitica1(xx, tempo)))
plt.grid()
plt.show()
```

# Schrodinger

```python
from scipy.constants import hbar, m_e
Nx = 1000
Nt = 1000
N = Nx - 2
L = 1e-8

xx = np.linspace(0, L, Nx)
dx = L / Nx
dt = 1e-18

alpha = hbar / (2 * m_e * dx**2)

mainA = (1 + dt * alpha * 1j) * np.ones(N)
offA = (-1j * dt * alpha / 2) * np.ones(N - 1)

mainB = (1 - dt * alpha * 1j) * np.ones(N)
offB = (1j * dt * alpha / 2) * np.ones(N - 1)

A = diags([mainA, offA, offA], offsets=[0, -1, 1], format='csc')
B = diags([mainB, offB, offB], offsets=[0, -1, 1], format='csc')

solveA = spla.factorized(A)

def psi0(x, sigma = 1e-10, x0 = L / 2, k = 5e10):
    return np.exp(- (x - x0)**2 / (2 * sigma**2)) * np.exp(1j * k * x)

psi = psi0(xx)
psi[0] = 0
psi[-1] = 0

fig, ax = plt.subplots()
line, = ax.plot(xx, np.abs(psi)**2)
ax.set_xlim(0, L)
ax.set_ylim(0, 1)
ax.grid()

def update(frame):
    global psi
    psi[1:-1] = solveA(B @ psi[1:-1])
    psi[0] = 0
    psi[-1] = 0
    line.set_ydata(np.abs(psi)**2)
    return line,

ani = FuncAnimation(fig, update, frames=Nt, interval=20, blit=True)
ani.save("Po√ßo_Inf1D.gif", writer=PillowWriter(fps=30))
plt.show()
```

# M√©todos Espectrais

$$
\psi_k = \sqrt{\frac{2}{L}} \,\sin\left(\frac{\pi k x}{L}\right)\, e^{-iEt/\hbar}
$$

$$
E = \frac{\pi^2 \hbar^2 k^2}{2 m L^2}
$$

$$
x_n = \frac{n L}{N}
$$

$$
\psi(x_n, t) = \sum_{k=1}^{N-1} b_k \sin\left(\frac{\pi k n}{N}\right) \exp\left(-i \frac{\pi^2 \hbar k^2}{2 m L^2} t \right)
$$

$$
\psi(x_n, 0) = \sum_{k=1}^{N-1} b_k \sin\left(\frac{\pi k n}{N}\right)
$$

$$
b_k = \alpha_k + i\eta_k
$$

```python
from scipy.fft import dst, idst

Nx = 1000
Nt = 500
N = Nx - 2

def psi0(x, sigma = 0.01, x0 = 0.5, k = 500):
    return np.exp(- (x - x0)**2 / (2 * sigma**2)) * np.exp(1j * k * x)

xx = np.linspace(0, 1, Nx)
psi0 = psi0(xx)
psi0[0] = 0
psi0[-1] = 0

alpha_k = dst(np.real(psi0[1:-1]), type=1)
eta_k = dst(np.imag(psi0[1:-1]), type=1)

bk = alpha_k + 1j * eta_k

k = np.arange(1, Nx-1)
A = np.pi**2 * k**2

fig, ax = plt.subplots()
line, = ax.plot(xx, np.abs(psi0)**2)
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.grid()

dt = 1e-6

def update(frame):
    t = frame * dt
    psi = np.zeros(Nx, complex)
    psi[1:-1] = idst(bk * np.exp(-1j * A * t), type=1)
    psi[0] = 0
    psi[-1] = 0
    line.set_ydata(np.abs(psi)**2)
    return line,

ani = FuncAnimation(fig, update, frames=Nt, interval=20, blit=True)
ani.save("Po√ßo_Inf1D_Adm_Spectral.gif", writer=PillowWriter(fps=30))
plt.show()
```
